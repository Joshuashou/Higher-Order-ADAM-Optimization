```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Training loss

Adam_4_train_accuracy = np.load('Simulation_Results/4_Moment_Adam/HAdam_train_accuracies.npy')

Vanilla_Adam_train_accuracy = np.load('Simulation_Results/Adam/Adam_train_accuracies.npy')

Adamax_train_accuracy = np.load('Simulation_Results/Adamax/Adamax_train_accuracies.npy')


Adam_4_test_accuracy = np.load('Simulation_Results/4_Moment_Adam/HAdam_test_accuracies.npy')

Vanilla_Adam_test_accuracy = np.load('Simulation_Results/Adam/Adam_test_accuracies.npy')

Adamax_test_accuracy = np.load('Simulation_Results/Adamax/Adamax_test_accuracies.npy')


```

```python
X = np.arange(15)

plt.plot(X, Adam_4_train_accuracy[:15], label='Higher Order Adam')
plt.plot(X, Adamax_train_accuracy[:15], label='Adamax')
plt.plot(X, Vanilla_Adam_train_accuracy[:15], label='Vanilla Adam')
plt.legend()
plt.title('RESNET CIFAR 10 Train Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.show()

plt.savefig("Simulation_Results/Train_Accuracies.png")
```

```python
X = np.arange(15)

plt.plot(X, Adam_4_test_accuracy[:15], label='Higher Order Adam')
plt.plot(X, Adamax_test_accuracy[:15], label='Adamax')
plt.plot(X, Vanilla_Adam_test_accuracy[:15], label='Vanilla Adam')
plt.legend()
plt.title('RESNET CIFAR 10 Test Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.show()

plt.savefig("Simulation_results/Test_Accuracies.png")
```
